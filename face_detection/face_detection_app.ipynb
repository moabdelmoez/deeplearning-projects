```python
####face detection with OpenCV####

#loading the image
test_image = cv2.imread("./image_1.jpeg")

#convert the test image to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

#display the image in gray color
plt.imshow(test_image_gray, cmap="gray")
```

```python
#function to convert the image from BGR to RGB
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#loading the classifier for frontal face

filename = "C:\\Users\\OsamaM2\\AppData\\Local\\Continuum\\anaconda3\\envs\\deeplearning\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml"
haar_cascade_face = cv2.CascadeClassifier(filename)
```

```python
haar_cascade_face
```

```python
"""
detectMultiscale module has two important parameters:
1- scalefactor: in group photo, it compensate the big faces that are near to the camera, and the ones that are far away
2- minNighbors: number of neighbors, a rectangle should have to be called
"""

face_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5)
```

```python
print("Face found: ", len(face_rects))
```

```python
#loop over all the coordinates it returned and draw rectangles around them using Open CV
for (x,y,w,h) in face_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
